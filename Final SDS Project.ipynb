```python
# Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

To conduct Exploratory Data Analysis (EDA) and apply descriptive statistics techniques to uncover insights regarding:

Distribution of educational programs across universities

Enrollment trends across levels of study

Popular disciplines and program types

Relationships between academic level, faculty, and program offerings

```python
# Load dataset
df = pd.read_csv(r"C:\Users\kumar\OneDrive\Desktop\Student_Enrollment_courses-2015.csv")
```

```python
# Quick Overview
print("Dataset Shape:", df.shape)
print("\nInfo:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
```

```python
# ========== UNIVARIATE ANALYSIS ==========
```

```python
# Summary Statistics (Numerical)
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Histogram: Programme ID
plt.figure(figsize=(8, 5))
sns.histplot(df['programme_id'], bins=30, kde=True)
plt.title("Distribution of Programme ID")
plt.xlabel("Programme ID")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot: Programme ID
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['programme_id'])
plt.title("Boxplot of Programme ID")
plt.show()
```

```python
# Countplot: Level
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='levell')
plt.title("Count of Students by Level")
plt.xlabel("Level")
plt.ylabel("Count")
plt.show()
```

```python
# Bar Plot ‚Äì Distribution of Program Types
plt.figure(figsize=(6,4))
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Distribution of Program Types')
plt.xlabel('Program Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Bar Plot ‚Äì Levels of Study
plt.figure(figsize=(6,4))
sns.countplot(x='levell', data=df, palette='Set3')
plt.title('Distribution of Levels of Study')
plt.xlabel('Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Top 10 Discipline Groups
plt.figure(figsize=(10,5))
top_disciplines = df['discipline_group'].value_counts().nlargest(10)
sns.barplot(x=top_disciplines.values, y=top_disciplines.index, palette='coolwarm')
plt.title('Top 10 Discipline Groups')
plt.xlabel('Count')
plt.ylabel('Discipline Group')
plt.tight_layout()
plt.show()
```

```python
# ========== BIVARIATE ANALYSIS ==========
```

```python
# Scatterplot: Programme ID vs Discipline Group ID
plt.figure(figsize=(7, 5))
sns.scatterplot(data=df, x='programme_id', y='discipline_group_id', hue='type')
plt.title("Programme vs Discipline Group")
plt.show()
```

```python
# Correlation Matrix
num_df = df.select_dtypes(include=['int64', 'float64'])
corr = num_df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()

# Cross-tab: Level vs Type
ct = pd.crosstab(df['levell'], df['type'])
print("\nCross-tabulation: Level vs Type")
print(ct)
```

```python
# Count Plot ‚Äì Level vs Program Type
plt.figure(figsize=(6,4))
sns.countplot(x='levell', hue='type', data=df, palette='Set1')
plt.title('Level of Study by Program Type')
plt.xlabel('Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Discipline Group vs Level (Top 10 Only)
plt.figure(figsize=(10,5))
top_disc = df['discipline_group'].value_counts().nlargest(10).index
sns.countplot(y='discipline_group', hue='levell', data=df[df['discipline_group'].isin(top_disc)], palette='Set2')
plt.title('Top Discipline Groups by Level')
plt.xlabel('Count')
plt.ylabel('Discipline Group')
plt.tight_layout()
plt.show()
```

```python
# Box Plot ‚Äì Programme ID by Level
plt.figure(figsize=(8,4))
sns.boxplot(x='levell', y='programme_id', data=df, palette='coolwarm')
plt.title('Programme ID Distribution by Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Cross Tab ‚Äì Level and Type
cross_tab = pd.crosstab(df['levell'], df['type'])
print("Cross Tabulation of Level and Program Type:\n")
print(cross_tab)
```

```python
# ========== MULTIVARIATE ANALYSIS ==========
```

```python
# Pairplot of numeric columns
sns.pairplot(num_df.sample(500), diag_kind='kde')  # Sample for performance
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()
```

```python
# Grouped Mean Analysis: Mean Programme ID by Level and Type
grouped = df.groupby(['levell', 'type'])['programme_id'].mean().unstack()
grouped.plot(kind='bar', figsize=(8, 5))
plt.title("Mean Programme ID by Level and Type")
plt.ylabel("Mean Programme ID")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Heatmap of Grouped Data
plt.figure(figsize=(6, 4))
sns.heatmap(grouped, annot=True, fmt=".1f", cmap="coolwarm")
plt.title("Heatmap: Programme ID by Level and Type")
plt.show()
```

```python
# Heatmap ‚Äì Correlation Matrix
plt.figure(figsize=(8,6))
numeric_cols = ['University_id', 'programme_id', 'discipline_group_id']
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Pair Plot ‚Äì Numeric Columns
sns.pairplot(df[numeric_cols], diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle("Pair Plot of Numeric Variables", y=1.02)
plt.show()
```

```python
# Grouped Bar Plot ‚Äì Level + Type + Discipline
grouped = df.groupby(['levell', 'type'])['discipline_group'].count().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=grouped, x='levell', y='discipline_group', hue='type', palette='magma')
plt.title('Discipline Group Count by Level and Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Stacked Bar ‚Äì University vs Level and Type (Top 5 Universities)
top_universities = df['University_name'].value_counts().nlargest(5).index
filtered_df = df[df['University_name'].isin(top_universities)]

pivot_table = pd.crosstab(index=filtered_df['University_name'], 
                          columns=[filtered_df['levell'], filtered_df['type']])

pivot_table.plot(kind='bar', stacked=True, figsize=(12,6), colormap='viridis')
plt.title('Enrollment by Level and Type ‚Äì Top 5 Universities')
plt.ylabel('Count')
plt.xlabel('University')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
